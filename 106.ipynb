```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("covid.csv")
df
```

```python
df = df.drop(columns=["ethnicity", "process", "exposure_yn", "current_status", "icu_yn"])
```

```python
df["race"] = df["race"].replace({"Missing": np.nan, "Unknown": np.nan})
df["hosp_yn"] = df["hosp_yn"].replace({"Missing": np.nan, "Unknown": np.nan})
df["death_yn"] = df["death_yn"].replace({"Missing": np.nan, "Unknown": np.nan})
df["age_group"] = df["age_group"].replace({"Missing": np.nan, "Unknown": np.nan})
df["sex"] = df["sex"].replace({"Missing": np.nan, "Unknown": np.nan})
df["underlying_conditions_yn"] = df["underlying_conditions_yn"].replace({"Missing": np.nan, "Unknown": np.nan})
df["symptom_status"] = df["symptom_status"].replace({"Missing": np.nan, "Unknown": np.nan})
df["case_positive_specimen_interval"] = df["case_positive_specimen_interval"].replace({np.nan: 0.0})
df["case_onset_interval"] = df["case_positive_specimen_interval"].replace({np.nan: 0.0})
df
```

```python
df = df.dropna(how="any")
df
```

```python
df = df.sort_values(by="case_month")
df
```

```python
df.to_csv('cleaning_data.csv', index = False)
```
